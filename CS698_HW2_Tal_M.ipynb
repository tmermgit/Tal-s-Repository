```python
# Step 1: Install Required Libraries
!pip install fiftyone detectron2 torch torchvision torchaudio pycocotools -q

# Step 2: Import Necessary Libraries
import fiftyone as fo
import fiftyone.zoo as foz
import os
import cv2
import numpy as np
import torch
import detectron2
from detectron2.utils.visualizer import Visualizer
from detectron2.data import MetadataCatalog, DatasetCatalog
from detectron2.engine import DefaultTrainer
from detectron2.config import get_cfg
from detectron2.evaluation import COCOEvaluator, inference_on_dataset
from detectron2.data import build_detection_test_loader
from detectron2.structures import BoxMode
import matplotlib.pyplot as plt

# Step 3: Load Dataset (Replace with Your Dataset)
dataset_name = "coco-2017"  # Change if using a different dataset
dataset = foz.load_zoo_dataset(dataset_name, split="validation")

# Step 4: Convert FiftyOne Dataset to Detectron2 Format
def get_fiftyone_dicts(fo_dataset):
    dataset_dicts = []
    for sample in fo_dataset:
        record = {
            "file_name": sample.filepath,
            "image_id": sample.id,
            "height": sample.metadata.height,
            "width": sample.metadata.width,
        }

        annotation_field = "ground_truth"  # Change if needed
        if not hasattr(sample, annotation_field) or sample[annotation_field] is None:
            print(f"Skipping sample {sample.id}: No annotations found")
            continue

        objs = []
        for detection in sample[annotation_field].detections:
            obj = {
                "bbox": detection.bounding_box,
                "bbox_mode": BoxMode.XYXY_ABS,
                "segmentation": getattr(detection, "mask", None),
                "category_id": detection.label,
            }
            objs.append(obj)

        record["annotations"] = objs
        dataset_dicts.append(record)

    return dataset_dicts

# Step 5: Register Dataset with Detectron2
dataset_name = "fiftyone_dataset"
if dataset_name in DatasetCatalog.list():
    DatasetCatalog.remove(dataset_name)
    MetadataCatalog.remove(dataset_name)

DatasetCatalog.register(dataset_name, lambda: get_fiftyone_dicts(dataset))
MetadataCatalog.get(dataset_name).set(thing_classes=dataset.default_classes)

metadata = MetadataCatalog.get(dataset_name)
dataset_dicts = get_fiftyone_dicts(dataset)

# Step 6: Configure Mask R-CNN Model
cfg = get_cfg()
cfg.merge_from_file("detectron2/configs/COCO-InstanceSegmentation/mask_rcnn_R_50_FPN_3x.yaml")
cfg.DATASETS.TRAIN = (dataset_name,)
cfg.DATASETS.TEST = ()
cfg.DATALOADER.NUM_WORKERS = 2
cfg.MODEL.WEIGHTS = "detectron2://ImageNetPretrained/MSRA/R-50.pkl"  # Pretrained weights
cfg.SOLVER.IMS_PER_BATCH = 2
cfg.SOLVER.BASE_LR = 0.00025
cfg.SOLVER.MAX_ITER = 1000  # Adjust based on dataset size
cfg.MODEL.ROI_HEADS.BATCH_SIZE_PER_IMAGE = 128
cfg.MODEL.ROI_HEADS.NUM_CLASSES = len(dataset.default_classes)

# Step 7: Train the Model
os.makedirs(cfg.OUTPUT_DIR, exist_ok=True)
trainer = DefaultTrainer(cfg)
trainer.resume_or_load(resume=False)
trainer.train()

# Step 8: Evaluate the Model
evaluator = COCOEvaluator(dataset_name, cfg, False, output_dir="./output/")
val_loader = build_detection_test_loader(cfg, dataset_name)
print(inference_on_dataset(trainer.model, val_loader, evaluator))

# Step 9: Save the Model
model_path = os.path.join(cfg.OUTPUT_DIR, "model_final.pth")
print(f"Model saved at: {model_path}")
```

Features of the Mask R-CNN Model.


ResNet-50 + FPN:
-Extracts multi-scale image features
-Improves detection of small objects

Region Proposal Network (RPN):
-Proposes likely object regions
-Reduces unnecessary computation

ROIAlign:
-Ensures precise object alignment
-Increases segmentation accuracy

Multi-task Head:
-Performs detection, classification, and masking
-Enables instance segmentation

Pretrained Weights:
-Uses ImageNet-trained backbone
-Faster and more accurate training

Custom Dataset Training:
-Adapts to new object categories
-Useful for domain-specific applications

COCO Evaluation Metrics:
-Measures model performance
-Enables objective model comparison
